```
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

      cos(3a)+cos(3b)

```
N=1000
X_train = np.random.random((N,2))*6-3    #  [-3,3]
y_train = np.cos(2*X_train[:,0])+np.cos(3*X_train[:,1])
```

```
fig =plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter(X_train[:,0],X_train[:,1],y_train)
plt.show()
```

```
model = tf.keras.Sequential([
          tf.keras.layers.Dense(128, input_shape=(2,),activation='relu'),
          tf.keras.layers.Dense(1),
])
```

```
opt = tf.keras.optimizers.Adam(learning_rate=0.2)
model.compile(optimizer=opt,loss='mse')
r= model.fit(X_train,y_train,epochs=100)
```

```
plt.plot(r.history['loss'],label='Loss')
```

```
fig =plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter(X_train[:,0],X_train[:,1],y_train)

line = np.linspace(-5,5,50)
X, Y = np.meshgrid(line, line)
X_grid = np.vstack((X.flatten(),Y.flatten())).T
yhat = model.predict(X_grid).flatten()
ax.plot_trisurf(X_grid[:,0],X_grid[:,1],yhat,linewidth=.2,antialiased=True)
plt.show()
```
